## Risk-Return Composite

##### We want funds with High returns and Low volatility

```python
# Normalize return and volatility
df['Return_1Y_Z'] = (df['Return_1Y_Pct'] - df['Return_1Y_Pct'].mean()) / df['Return_1Y_Pct'].std()
df['Volatility_1Y_Z'] = (df['Volatility_1Y_Pct'] - df['Volatility_1Y_Pct'].mean()) / df['Volatility_1Y_Pct'].std()

# Higher is better
df['Risk_Return_Score'] = df['Return_1Y_Z'] - df['Volatility_1Y_Z']

df[['Risk_Return_Score']].describe()
```

# Model 1: Return Model

```python
model_df = df.copy()

# Log-transform AUM
model_df["Log_AUM"] = np.log(model_df["Fund_Net_Assets_USD_M"])

# Encode Fund Type (ETF = 1, Mutual Fund = 0)
model_df["ETF_Dummy"] = (model_df["Fund_Type"] == "ETF").astype(int)

# Create sector and geography dummies
sector_dummies = pd.get_dummies(model_df["Primary_Sector"], prefix="Sector", drop_first=True)
geo_dummies = pd.get_dummies(model_df["Geographic_Focus"], prefix="Geo", drop_first=True)

model_df = pd.concat([model_df, sector_dummies, geo_dummies], axis=1)
```

```python
import statsmodels.api as sm

X_vars = [
    "Tariff_Risk_Score",
    "Trade_War_Resilience_Score",
    "Volatility_1Y_Pct",
    "Expense_Ratio_Pct",
    "Log_AUM",
    "ETF_Dummy"
]

X = model_df[X_vars + list(sector_dummies.columns) + list(geo_dummies.columns)]
y = model_df["Return_1Y_Pct"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

model.summary()
```

Big picture: Is this model good?
No, not really  and that itself is an important insight.

Heres why:

Model fit

R-squared = 0.156

The model explains only ~16% of the variation in 1-year returns

Adjusted R-squared = 0.033

After penalizing for too many variables, explanatory power drops to ~3%

 Interpretation

Short-term (1Y) fund returns are largely driven by market noise and macro shocks, not easily explained by static fund characteristics alone.

This is very realistic for financial data.

Overall model significance

Prob(F-statistic) = 0.145 (> 0.05)

 This means:

Collectively, the variables do not strongly explain 1-year returns at conventional significance levels.

Again  this is common in asset return models and does not invalidate your analysis.

 What do the key variables say?

Now lets interpret the important ones, not every dummy.

 Tariff_Risk_Score

Coefficient: +0.05

p-value: 0.78  (not significant)

 Interpretation:

Tariff risk by itself does NOT explain 1-year returns in a statistically reliable way.

Why this makes sense

Tariffs affect firms over multi-year horizons

1Y returns are dominated by:

sentiment

interest rates

global risk-on/risk-off moves

 This supports a longer investment horizon later.

 Trade_War_Resilience_Score

Coefficient: +0.22

p-value: 0.15 (borderline, but not significant)

 Interpretation:

More resilient funds tend to have higher returns, but the effect is not strong enough to be conclusive in 1 year.

Important nuance:

Weak significance  useless insight

 Volatility_1Y_Pct

Coefficient: +0.08

p-value: 0.57

 This is counterintuitive at first, but:

Short-term regressions can flip signs

This is why volatility is better for risk discussion than return prediction.

 Expense_Ratio, AUM, ETF dummy

All insignificant

 Interpretation:

Cost structure and size do not meaningfully explain short-term performance under tariff stress.

 Sector & geography dummies  any signals?

Most are insignificant  thats expected with 38 variables.

But note these near-significant ones (p  0.050.10):

4 Diagnostics  is the regression technically okay?
Residuals

Omnibus / JB p-values > 0.7
 Residuals are approximately normal

 No autocorrelation issue

This is important.

There is multicollinearity, mainly from:

Geographic dummies

Highly correlated exposure variables

This explains:

Inflated standard errors

Tariff exposure and resilience do not explain short-term returns well  but they matter for risk control, drawdowns, and medium-to-long-term outcomes.

# Model 2: Risk Model (Volatility)

```python
import statsmodels.api as sm

# Dependent variable (Risk)
y_risk = model_df['Volatility_1Y_Pct']

# Independent variables

X= [
    "Tariff_Risk_Score",
    "Trade_War_Resilience_Score",
    "Supply_Chain_Vulnerability_Score",
    "Expense_Ratio_Pct",
    "Log_AUM",
    "ETF_Dummy"
]
X=model_df[X]
# Add intercept
X = sm.add_constant(X)

# Fit model
model_risk = sm.OLS(y_risk, X).fit()

# Output summary
print(model_risk.summary())
```

#### How good is this model overall?
Model fit

R-squared  0.16

Adjusted R-squared  0.03

Prob(F-statistic) = 0.145


The model explains only a small portion of variation in risk, and the overall model is not statistically strong.

This is not a failure.

 In finance, especially with cross-sectional fund data:

Risk drivers are diffuse

Many factors are unobservable

Multicollinearity weakens significance

 Interpretation of key variables (this is what matters)
 Tariff_Risk_Score

Higher tariff exposure does not mechanically increase short-term volatility across funds.
